## Labels are the binary/fake flag - real (0), fake 1

```python
labels = getBinaryLabels()
```

```python
print('Total Records:',len(labels))
print('Total Fake: ',sum(labels))
print('First 10 labels',labels[:10])
```

```python
# Input parameters to shape data
vocab_size = 5000 # Total Vocab size - anything bigger is rare
seq_len = 40 # Number of Words per record
# trn_data includes all data
trn_data = formatTrainingData(getTrainingData(),vocab_size,seq_len)
```

## Show the first record of our Training Data

```python
print(trn_data[1])
```

```python
factor = 0.80  # How much data is used for training vs validation

t_index = round(factor*len(labels))
print (t_index)

model = createBinaryModel(len(trn_data[0]), vocab_size)
```

```python
model.fit(trn_data[:t_index], labels[:t_index], validation_data=(trn_data[t_index:len(trn_data)], labels[t_index:len(trn_data)]), epochs=4, batch_size=256)
```

```python
model.save('./Models/FakeNews-v7.h5')
```

```python
preds = model.predict_classes(trn_data[t_index:len(trn_data)])
f1_score(labels[t_index:len(trn_data)],preds,labels=list(set(labels)),average=None)
```

```python
confusion_matrix(labels[t_index:len(trn_data)],preds)
```
